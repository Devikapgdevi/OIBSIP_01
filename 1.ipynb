```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load the retail sales dataset from a CSV file
file_path = "Desktop\data analytics\EDA on reatil sale data.csv"  # Replace with your CSV file path
df = pd.read_csv('retail_sales_dataset.csv')

# Display the first few rows of the dataset to understand its structure
print("Loaded Data:")
print(df.head())

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python


# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Optional: Drop rows with missing values
df.dropna(inplace=True)  # Uncomment if you want to drop rows with missing values
```

```python
# Descriptive Statistics
print("\nDescriptive Statistics:")
print(df.describe(include='all'))
```

```python
# Set the 'Date' column as the index for time series analysis
df.set_index('Date', inplace=True)

# Resample the data to get monthly sales totals
monthly_sales = df.resample('M').sum()
```

```python
# Plotting Monthly Total Sales
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales['Total Amount'], marker='o')
plt.title('Monthly Total Sales')
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Customer Analysis
# Analyze gender distribution of customers
gender_distribution = df['Gender'].value_counts()

# Plotting Gender Distribution
plt.figure(figsize=(5, 5))
sns.barplot(x=gender_distribution.index, y=gender_distribution.values, palette='pastel')
plt.title('Customer Gender Distribution')
plt.ylabel('Count')
plt.xlabel('Gender')
plt.tight_layout()
plt.show()
```

```python
# Analyze Age Distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Age'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Product Category Analysis: Total sales by product category
product_sales = df.groupby('Product Category')['Total Amount'].sum()

# Plotting Total Sales by Product Category
plt.figure(figsize=(10, 5))
sns.barplot(x=product_sales.index, y=product_sales.values, palette='muted')
plt.title('Total Sales by Product Category')
plt.ylabel('Total Amount')
plt.xlabel('Product Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
